```python
#Question4: Write a python program to create two threads. Thread one must print the list of squares and thread
# two must print the list of cubes.


import threading

def print_squares():
    for i in range(1, 11):
        print(i * i)

def print_cubes():
    for i in range(1, 11):
         print(i * i * i)

if __name__ == '__main__':
    t1 = threading.Thread(target=print_squares)
    t2 = threading.Thread(target=print_cubes)
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()

    print("Done!")
```

```python
#Question5: State advantages and disadvantages of multithreading.




"""Multithreading is a powerful technique that enables programs to perform multiple tasks simultaneously. However, it also has 
both
advantages and disadvantages that should
be considered when deciding whether to use it in a particular application. Here are some of the advantages and disadvantages of
multithreading:

Advantages:

Improved performance: Multithreading can increase the efficiency and speed of a program by taking advantage of multiple CPU 
    cores and allowing tasks to run in parallel.

Resource sharing: Multithreading allows multiple threads to access shared resources, such as memory or I/O devices, without 
    interfering with each other.

Responsiveness: Multithreading can make a program more responsive to user input by allowing tasks to run in the background 
    while the user interacts with the program's user interface.

Simplified design: Multithreading can simplify the design of a program by allowing it to be divided into smaller, more
    manageable tasks that can be executed concurrently.
    
    
Disadvantages:

Complexity: Multithreaded programs are generally more complex than single-threaded programs and can be harder to develop, 
    debug, and maintain.

Synchronization issues: Multithreading can introduce synchronization issues, such as race conditions and deadlocks, that 
    must be carefully managed to ensure correct program behavior.

Overhead: Multithreading can introduce overhead, such as the cost of creating and managing threads and the cost of 
    synchronization, which can reduce the overall performance of a program.

Resource contention: Multithreading can lead to resource contention, where multiple threads are competing for the same 
    resources, which can result in reduced performance or even program crashes.    """
```

```python
#Question6: Explain deadlocks and race conditions.




"""Deadlocks and race conditions are two types of synchronization issues that can arise in multithreaded programs.

Deadlocks: A deadlock occurs when two or more threads are blocked, waiting for each other to release resources that they are 
    holding. In other words, a deadlock happens when two threads hold locks on resources that the other needs to proceed, and 
    neither thread can release its lock until it gets the lock it needs. As a result, the threads are 
    stuck in a circular waiting state, and the program is unable to make any progress.

    
Deadlocks can be difficult to detect and resolve, as they can occur unpredictably and sporadically. To prevent deadlocks, it is 
essential to ensure that threads acquire locks on resources in a consistent order and release them when they are no longer 
needed.

Race conditions: A race condition occurs when two or more threads access a shared resource concurrently and the final result
    of the operation depends on the timing and order of the thread execution. In other words, the behavior of the program 
    depends on the relative timing of the threads, and it can change unpredictably from one run to another.
    
    
    
Race conditions can cause the program to produce incorrect results, crash, or exhibit other unexpected behavior. To prevent 
race conditions, it is necessary to use synchronization mechanisms, such as locks or semaphores, to control access to shared
resources 
and ensure that only one thread accesses the resource at a time.    """
```
